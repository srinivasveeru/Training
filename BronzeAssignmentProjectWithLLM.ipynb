```python
#Use case: “Policy & Claims Copilot” (Customer support + Claims pre-check) using RAG
#Steps cover Loading document, Creating Chunks, Embedding, and storing in vector database
!pip install openai pypdf

import os
from pypdf import PdfReader
from openai import OpenAI

#API Key to be obtained from secret key in Google colab
from google.colab import userdata
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
# --- Configuration ---
PDF_PATH = "InsurancePolicy.pdf"

# Ensure the PDF file exists
if not os.path.exists(PDF_PATH):
    print(f"Error: {PDF_PATH} not found. Please add the PDF file.")
    exit()

# 1. Extract text from the PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    reader = PdfReader(pdf_path)
    for page in reader.pages:
        text += page.extract_text() or ""
    return text

pdf_content = extract_text_from_pdf(PDF_PATH)

#code created for any queries from user pertaining to the document
USER_QUERY = input()

# Send the prompt to the LLM and get the answer
try:
    client = OpenAI()
    response = client.chat.completions.create(
        model="gpt-4o-mini", # using free model
        messages=[
            {"role": "user", "content": USER_QUERY}
        ],
        max_tokens=500,
        temperature=0.1
    )
    print("--- Plain LLM Response ---")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"An error occurred: {e}")
```
