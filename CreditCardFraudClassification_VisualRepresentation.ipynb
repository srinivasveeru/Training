```python
!pip install tensorflow pydot graphviz pandas scikit-learn

import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OrdinalEncoder
import os
from tensorflow.keras.utils import plot_model

# Suppress warnings and set TensorFlow logging level
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import warnings
warnings.filterwarnings('ignore')
```

```python
def fraud_detection_model(input_dim):
    """Builds a simple sequential neural network model."""
    model = Sequential([
        Dense(64, activation='relu', input_shape=(input_dim,)),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid') # Sigmoid for binary classification (fraud or not)
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

def visualize_model(model, filename="neural_network_architecture.png"):
    """Visualizes the neural network architecture and saves it to a file."""
    try:
        plot_model(model, to_file= filename, show_shapes=True, show_layer_names=True, show_layer_activations=True)
        print(f"Neural network architecture visualization saved to {filename}")
    except ImportError:
        print("Error: 'pydot' or 'graphviz' not found. Please ensure they are installed and Graphviz is in your system PATH.")
    except Exception as e:
        print(f"An error occurred during visualization: {e}")

# --- Main execution ---
if __name__ == "__main__":

    try:
        #csv file taken from Kaggle and take required columns alone
        df = pd.read_csv('CardTxnsBronze.csv')
        num_cols = ['Transaction_Amount','Customer_Age','Account_Balance']
        str_cols = ['Transaction_Date', 'Category']
        X = df.drop(['Transaction_ID','Fraud_Flag'],axis=1)
        y = df['Fraud_Flag']

        # Standardize features (important for NNs)
        scaler = StandardScaler()
        X[num_cols] = scaler.fit_transform(X[num_cols])

        for col in str_cols:
          le = OrdinalEncoder()
          X[str_cols] = le.fit_transform(X[str_cols])

        #print(X)
        # Build the model
        input_dim = X.shape[1]
        model = fraud_detection_model(input_dim)
        model.summary()

        # Visualize the model
        visualize_model(model)

    except FileNotFoundError:
        print("Dataset file not found. Please ensure 'CardTxnsBronze.csv' is in the current directory or provide the correct path.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
